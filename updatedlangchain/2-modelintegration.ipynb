## Models Integration with OpenAI , Google Gemini and GROQ

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GEMINI_API_KEY"]=os.getenv("GEMINI_API_KEY")

```

```python
from langchain.chat_models import init_chat_model
model= init_chat_model("google_genai:gemini-2.5-flash")
response=model.invoke("Why do parrots fly?")
response.content
```

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model= init_chat_model("groq:qwen/qwen3-32b")
response=model.invoke("Why do parrots fly?")
response.content
```

```python
import os
from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model= ChatGroq(model="qwen/qwen3-32b")
response=model.invoke("Why do parrots fly?")
response.content
```

### Streaming and Batch

```python
### Streaming is when the LLM doesnt wait for the answers,and starts streaming immediately
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="" , flush=True)
```

```python
responses=model.batch(["Why do parrots have colorful feathers?", 
                      "Why is sky blue?",
                      "Why do you get tears even when you are happy?"],
                      max_concurrency=5)
for response in responses:
    print(response.content)
```
